# Matplotlib tutorial
Matplotlib is used to visulize data in python

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Plotting points on graph

```python
# generate 2 numpy arrays of size 1000 where each value is radom between 1 and 100
X_data = np.random.random(1000) * 100 
Y_data = np.random.random(1000)*100

plt.scatter(X_data, Y_data)
plt.show()
```

now let's do some customizations in the graph.
like, 
1. set colour of pointer (circle) to green (#0f0). 
2. increase the size of pointer to 150, set marker shape to * instead of a circle.
3. finally set it's visibility to 0.3 i.e., make it a bit transparent

```python
plt.scatter(X_data, Y_data, c='#0f0', s=150, marker='*', alpha=0.3)
plt.show()
```

### 2. Plotting a continuous graph
the customization are:
1. colour of line is red
2. width of the line (lw) is now 3
3. instead of a continuous line, we choose dashed line (--)

```python
years = [2010 + x for x in range(15)]
weights = np.random.normal(60, 10, 15)

plt.plot(years, weights, c='#f00', lw=3, linestyle='--')
plt.show()
```

### 3. Bar plots
customizations are:
1. color of bar is red (**here we use 'color' for color instead of just 'c'**)
2. width of bar is 0.8
3. border width of 5 (lw)
4. border color of green
5. aligniment of bar to the edge of label. by default the label (x) would be in the middle of bar.

```python
x = ['c++', 'python', 'R', 'SQL']
y = [23, 45, 9, 78]

plt.bar(x, y, color='#f00', width=0.8, lw=5, edgecolor='green', align='edge')
plt.show()
```

### 5. Pie charts

```python
langs = ['python', 'c++', 'c#', 'SQL', 'java']
votes = [432, 293, 89, 120, 378]

plt.pie(votes, labels=langs)
plt.show()
```

customization: let's say we want the python part of pie to explode out of the whole pie

```python
explodes = [0.1, 0, 0, 0, 0]
plt.pie(votes, labels=langs, explode=explodes)
plt.show()
```

now, let's say we want to show percent share of each language  
here: '%.2f' means show just value of percent upto 2 decimals, like 32.21.  
while '%.2f%%' means, show percent with % sign, like 32.21%

pctdistance=0.5 here means the distance of percent text from center of pie.

```python
plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%', pctdistance=0.5)
plt.show()
```

## Plot customization
here we are adding:
1. title of plot
2. xlabel: label for values in x axis
3. ylabel
4. yticks: list of values that should be shown in y axis. and also any extra text with those values. like their unit or currency.

```python
years = [2010 + x for x in range(15)]
weights = np.random.normal(60, 10, 15)
yticks = list(range(40, 80, 5))

plt.plot(years, weights)
plt.title("Weight of John over year")
plt.xlabel("Years")
plt.ylabel("Weightd")
plt.yticks(yticks, [f'{x} kg' for x in yticks])
plt.show()
```

now, you can also customize font for each texts (whether title or label). like font size, font family

```python
plt.plot(years, weights)
plt.title("Weight of John over year", fontsize=30, fontname='Algerian')
plt.xlabel("Years")
plt.ylabel("Weightd")
plt.yticks(yticks, [f'{x} kg' for x in yticks])
plt.show()
```

### Multiple plots
we can also add multiple plots in one single graph.

```python
stockA = [123, 122, 112, 90, 140]
stockB = [45, 50, 46, 62, 90]
stockC = [150, 130, 129, 132, 120]

plt.plot(stockA, label="Google")
plt.plot(stockB, label='NVIDIA')
plt.plot(stockC, label="Apple")

plt.show()
```

now, to show a legend stating which line is of which company  
(optional) we can also specify location of legend windows by using loc param

```python
stockA = [123, 122, 112, 90, 140]
stockB = [45, 50, 46, 62, 90]
stockC = [150, 130, 129, 132, 120]

plt.plot(stockA, label="Google")
plt.plot(stockB, label='NVIDIA')
plt.plot(stockC, label="Apple")

plt.legend(loc="lower center")

plt.show()
```

#### You cab show same kind of legend window for pie charts also  
instead of labels on pie itself

```python
votes = [10, 2, 14, 8]
parties = ["a" "b", "c", "d"]

plt.pie(votes, labels=None)
plt.legend(labels=parties)
plt.show()
```

### Styling
we can also use different types of styling in our graphs.  
to do that we need to import style from matplotlib library and use pre defined styles

```python
from matplotlib import style
style.use("dark_background")
votes = [10, 2, 14, 8]
parties = ["a" "b", "c", "d"]

plt.pie(votes, labels=None)
plt.legend(labels=parties)
plt.show()
```

### Multiple windows
here's how we can make multiple different windows of graphs

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

plt.figure(1) #defining first window
plt.scatter(x1, y1)

plt.figure(2) #defining second window
plt.plot(x2, y2)

plt.show()
```

### Subplots
multiple plots in one window

```python
x = np.arange(40)

fig, axs = plt.subplots(2, 2) # creata a 2x2 grid of 4 subplots in one window

# acceessing top left subplot
axs[0, 0].plot(x, np.sin(x)) 
axs[0, 0].set_title("Sine wave")

# acceessing top right subplot
axs[0, 1].plot(x, np.log(x))
axs[0, 1].set_title("Log function")

# acceessing bottom left subplot
axs[1, 0].plot(x, np.cos(x))
axs[1, 0].set_title("Cosine wave")

# acceessing bottom right subplot
axs[1, 1].plot(x, np.random.random(40))
axs[1, 1].set_title("Random function")

# giving a super title for the window (i.e., all 4 subplots)
fig.suptitle("Four plots")

#give widthspcae and heightspace between subplots
plt.subplots_adjust(wspace=0.5, hspace=0.5)
# or we can also simply use plt.tight_layout() to give auto margin so that nothing overlaps

plt.show()
```

this way, we also customize each subplot just like we did with normal plots.

### Exporting plot figure
we can export the plot with some customizations lie,
1. setting dpi=300, higher=better resolution
2. transparency=False (default), sets background to true or false

```python
x = np.arange(40)

fig, axs = plt.subplots(2, 2) # creata a 2x2 grid of 4 subplots in one window

axs[0, 0].plot(x, np.sin(x)) 
axs[0, 0].set_title("Sine wave")

axs[0, 1].plot(x, np.log(x))
axs[0, 1].set_title("Log function")

axs[1, 0].plot(x, np.cos(x))
axs[1, 0].set_title("Cosine wave")

axs[1, 1].plot(x, np.random.random(40))
axs[1, 1].set_title("Random function")

fig.suptitle("Four plots")

plt.tight_layout()

plt.savefig("Exampleplot.png", dpi=300, transparent=True)
```
